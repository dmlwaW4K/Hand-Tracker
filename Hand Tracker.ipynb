```python
import cv2
from cvzone.HandTrackingModule import HandDetector
```

```python
cap = cv2.VideoCapture(0)
```

```python
detector = HandDetector(detectionCon=0.8, maxHands=2)

while True:
    # Capture frame-by-frame
    success, img = cap.read()
    if not success:
        break

    # Detect hands
    hands, img = detector.findHands(img)

    # Process each hand found
    if hands:
        for hand in hands:
            # Get the landmarks
            lmList = hand["lmList"]  # List of 21 landmarks points
            bbox = hand["bbox"]  # Bounding box info x,y,w,h
            centerPoint = hand['center']  # Center of the hand cx,cy
            handType = hand["type"]  # Hand type Left or Right

            # Draw a circle on the center of the hand
            cv2.circle(img, centerPoint, 10, (255, 0, 255), cv2.FILLED)

            # Optionally, you can draw landmarks and bbox
            # for lm in lmList:
            #     cv2.circle(img, (lm[0], lm[1]), 5, (0, 255, 0), cv2.FILLED)
            cv2.rectangle(img, (bbox[0], bbox[1]), (bbox[0] + bbox[2], bbox[1] + bbox[3]), (0, 255, 0), 2)
            cv2.putText(img, handType, (bbox[0], bbox[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 255), 2)

    # Display the resulting frame
    cv2.imshow('Hand Tracking', img)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
